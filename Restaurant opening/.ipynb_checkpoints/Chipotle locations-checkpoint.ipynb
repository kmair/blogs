### Neighborhood venues for analysis

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    fail_list = []
    venues_list=[]
    for name, lat, lng in tqdm(zip(names, latitudes, longitudes)):     
        
        try:
            # create the API request URL
            url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
                CLIENT_ID, 
                CLIENT_SECRET, 
                VERSION, 
                lat, 
                lng, 
                radius, 
                LIMIT)

            # make the GET request
            results = requests.get(url).json()["response"]['groups'][0]['items']

            # return only relevant information for each nearby venue
            venues_list.append([(
                name, 
                lat, 
                lng, 
                v['venue']['name'], 
                v['venue']['location']['lat'], 
                v['venue']['location']['lng'],  
                v['venue']['categories'][0]['name']) for v in results])
            
        except:
            fail_list.append(name)
            
    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return nearby_venues, fail_list
```

### Neighborhood analysis

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row#.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]

def most_common_venues(freq_data):
    num_top_venues = 10

    indicators = ['st', 'nd', 'rd']

    # create columns according to number of top venues
    columns = ['Neighborhood', 'Location_tag']
    for ind in np.arange(num_top_venues):
        try:
            columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
        except:
            columns.append('{}th Most Common Venue'.format(ind+1))

    # create a new dataframe
    neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
    neighborhoods_venues_sorted['Neighborhood'] = freq_data['Neighborhood']
    neighborhoods_venues_sorted['Location_tag'] = freq_data['Location_tag']

    freq_data = freq_data.drop(columns=['Neighborhood', 'Location_tag'])
    
    for ind in np.arange(freq_data.shape[0]):
        neighborhoods_venues_sorted.iloc[ind, 2:] = return_most_common_venues(freq_data.iloc[ind, :], 
                                                                              num_top_venues)

    return neighborhoods_venues_sorted
```

```python
test_venues_sorted = most_common_venues(test)
train_venues_sorted = most_common_venues(train)
```

```python
from collections import Counter

def plot_top(common_values, limit=20):
    cnt = Counter(common_values.flatten())
    cnt_sorted = {k: v for k, v in sorted(cnt.items(), key=lambda item: item[1], reverse=True)}

    labels, count = zip(*cnt_sorted.items())
    labels = labels[:limit]
    count = count[:limit]
    indexes = np.arange(len(labels))
    width = 1

    fig, ax = plt.subplots()

    # Example data
    y_pos = np.arange(len(count))

    ax.barh(y_pos, count, align='center', color='#bc5090')
    ax.set_yticks(y_pos)
    ax.set_yticklabels(labels)
    ax.invert_yaxis()  # labels read top-to-bottom
    ax.set_xlabel('Frequency')
    ax.set_title('Occurrence in top 10 venues')

    plt.show()
```

### Selecting the K

```python
train_clustering = train.drop(columns=['Neighborhood', 'Location_tag'])
```

```python
def best_k(X, min_k, max_k):
    # Taking minimum of 4 clusters
    # set number of clusters
    distortions, inertia = [], []

    n_clusters = np.arange(min_k, max_k+1)
    for k in tqdm(n_clusters):
        kmeans = KMeans(n_clusters=k).fit(X)
        
        df = pd.DataFrame({'cluster count': np.bincount(kmeans.labels_)})
        
        print(df)
        print()

        distortions.append(sum(np.min(cdist(train_clustering, kmeans.cluster_centers_, 
                      'euclidean'),axis=1)) / train_clustering.shape[0]) 

        inertia.append(kmeans.inertia_)
        
    fig,ax = plt.subplots(1,2, figsize=(20,10))
    
    ax[0].plot(n_clusters, distortions)
    ax[0].set_title('Distortions')
    ax[1].plot(n_clusters, inertia, color='g')
    ax[1].set_title('Inertia')

# best_k(train_clustering, min_k = 10, max_k=20)
```

*This will give us 1 main cluster and remaining would contain the most common venue.*

### Clustering into 5 groups

```python
def common_venues(data, k=5):
    kmeans = KMeans(n_clusters=k).fit(data)
    cnt = np.bincount(kmeans.labels_)    
    categ = np.argsort(cnt)[0:-1] # Remove main cluster 

    print('Most common venues are:')

    for i, c in enumerate(categ[::-1]):
        ind = np.where(kmeans.cluster_centers_[c]>=0.99)
        print(f'{i+1}. {train_clustering.columns[ind][0]}')
```

- Main categories in top Chipotle locations

```python
train_cluster_top = train[train['Location_tag']=='top'].drop(columns=['Neighborhood', 'Location_tag'])
common_venues(train_cluster_top, k=11)
```

- Main categories in all other locations

```python
train_cluster_other = train[train['Location_tag']=='other'].drop(columns=['Neighborhood', 'Location_tag'])
common_venues(train_cluster_other, k=11)
```

In the above results, we can find that:
- The top locations are situated near normal businesses and restaturants
- **But, the top Chipotle locations are situated near various restaurant cuisines while the other locations commonly have other competing Mexican restaurants nearby**

We can draw the conclusion that in order to be a successful Chipotle franchise, it's better to operate in locations where there are no Mexican outlets nearby.

```python
mex_names = list(mex_df.name)
```

```python
venue_results
```

```python
def rest_details(venues):
    lat, lng, name, chip = [], [], [], []
    for venue in venues:
        n = venue['name']
        if n in mex_names:
            continue

        lat.append(venue['location']['lat'])
        lng.append(venue['location']['lng'])
        name.append(n)
                    
    return pd.DataFrame({'lat':lat, 'lng':lng, 'name':name})

rest_df = rest_details(venue_results)
```
